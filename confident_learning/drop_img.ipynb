```python
data_dir = "/workspace/yolo_dataset_4_dec"
```

```python
from pathlib import Path

data_root = Path(data_dir)
img_root = data_root / "images"
lbl_root = data_root / "labels"

exts = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}

def count_images(root: Path) -> int:
    return sum(1 for p in root.rglob("*") if p.is_file() and p.suffix.lower() in exts)

before = count_images(img_root)
print(f"Images before: {before}")

to_drop_set = set(to_drop)
dropped_imgs = 0
dropped_lbls = 0

for img_name in sorted(to_drop_set):
    stem = Path(img_name).stem
    
    for img_path in img_root.rglob(img_name):
        if img_path.is_file():
            img_path.unlink()
            dropped_imgs += 1
    
    for lbl_path in lbl_root.rglob(f"{stem}.txt"):
        if lbl_path.is_file():
            lbl_path.unlink()
            dropped_lbls += 1

after = count_images(img_root)
print(f"Dropped images: {dropped_imgs}")
print(f"Dropped labels: {dropped_lbls}")
print(f"Images after:  {after}")
```
